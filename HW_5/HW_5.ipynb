```python
import networkx as nx
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline
```

```python
G = nx.Graph()
edge_index = [(0, 1), (0, 2), (1, 2), (2, 3), (3, 4), (3, 5), (4, 5)]
G.add_edges_from(edge_index)
Q = nx.algorithms.community.modularity(G, [{0, 1, 2, 3, 4, 5}])
print('The Q of the following graph is : %f'%Q)
pic = mpimg.imread('pic/pic_1.png')
plt.imshow(pic)
plt.axis('off')
plt.show()
```
